```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from sklearn.model_selection import train_test_split,GridSearchCV,StratifiedKFold
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report 
import pandas as pd
```

```python
df = pd.read_csv("diabetes.csv")
df.head()
```

```python
df['Outcome'].value_counts()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
X = df.drop(['Outcome'],axis=1)
y = df['Outcome']
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25,stratify=y)
```

```python
model = LogisticRegression()
penalty = ['l1', 'l2', 'elasticnet']
C_score = [100,10,1,0.1]
solver = ['lbfgs', 'liblinear', 'newton-cg', 'newton-cholesky', 'sag', 'saga']
class_weight = [{0:w,1:v} for w in [1,10,50,100] for v in [1,10,50,100]]
```

```python
class_weight
```

```python
params = dict(penalty=penalty,C=C_score,solver=solver,class_weight=class_weight)
```

```python
params
```

```python
cv = StratifiedKFold(5)
```

```python
gridsearch = GridSearchCV(estimator=model,param_grid=params,cv=cv,scoring='accuracy')
```

```python
gridsearch.fit(X_train,y_train)
```

```python
print(gridsearch.best_params_)
```

```python
y_pred = gridsearch.predict(X_test)
```

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
cr = classification_report(y_test,y_pred)
print(cr)
```
